## 포켓몬 구현하기

> 피카츄를 클래스-인스턴스로 구현해 봅시다. 게임을 만든다면 아래와 같이 먼저 기획을 하고 코드로 구현하게 됩니다.
우선 아래와 같이 구현해 보고, 추가로 본인이 원하는 대로 구현 및 수정해 봅시다.

모든 피카츄는 다음과 같은 속성을 갖습니다.
* `name`: 이름
* `level`: 레벨
    * 레벨은 시작할 때 모두 5 입니다.
* `hp`: 체력
    * 체력은 `level` * 20 입니다.
* `exp`: 경험치
    * 상대방을 쓰러뜨리면 상대방 `level` * 15 를 획득합니다.
    * 경험치는 `level` * 100 이 되면, 레벨이 하나 올라가고 0부터 추가 됩니다. 

모든 피카츄는 다음과 같은 행동(메서드)을 할 수 있습니다.
* `bark()`: 울기. `'pikachu'` 를 출력합니다.
* `body_attack()`: 몸통박치기. 상대방의 hp 를 내 `level` * 5 만큼 차감합니다.
* `thousond_volt()`: 십만볼트. 상대방의 hp 를 내 `level` * 7 만큼 차감합니다.

```python
# 아래에 코드를 작성해주세요.
import random

class Pikachu:
    def __init__(self, name):
        self.name = name
        self.level = 5
        self.hp = self.level * random.randint(15,20)
        self.exp = 0
        print(f'{self.name}의 HP가 {self.hp}로 설정되었습니다.')
        self.skill= [20,0,0,3]
        
    def body_attack(self,defence):
        self.skill[0] -= 1
        if self.skill[0] < 0:
            print('스킬을 사용할 수 없습니다.')
            return None
        else:
            print(f'q스킬 {self.skill[0]}회남음')
        print(f'{self.name}이(가) 몸통박치기를 {defence.name}한테 몸통박치기!')
        a = random.random()
        b= self.level * random.randint(2,3)
        if a < 0.15:
            print(f'{self.name}의 몸통박치기가 빗나감')
            defence.hp -= 0
            print(f'{defence.name}의 체력이 {defence.hp}으로 그대로')
        elif a > 0.85:
            print(f'{self.name}의 몸통박치기가 급소에 맞았다!')
            defence.hp -= b * 2
            print(f'{defence.name}의 체력이 {defence.hp}가 되었다(hp{-b *2})')
        else:
            defence.hp -= b
            print(f'{defence.name}의 체력이 {defence.hp}가 되었다(hp{-b})')
        if defence.hp <= 0 :
            self.exp += 15
            print(f'{defence.name} 으앙주금')
            print(f'{self.name}이(가) + 15 경험치(현재 경험치 : {self.exp})')
        if self.exp >= 100:
            self.exp -= 100
            self.level += 1
            
    def thousond_volt(self,defence):
        self.skill[3] -= 1
        if self.skill[3] < 0:
            print('스킬을 사용할 수 없습니다.')
            return None
        else:
            print(f'r스킬 {self.skill[3]}회남음')
        print(f'{self.name}이(가) {defence.name}한테 십만볼트!')
        a = random.random()
        b= self.level * random.randint(3,5)
        if a < 0.2:
            print(f'{self.name}의 십만볼트가 빗나감')
            defence.hp -= 0
            print(f'{defence.name}의 체력이 {defence.hp}으로 그대로')
        elif a > 0.85:
            print(f'{self.name}의 십만볼트가 급소에 맞았다!')
            defence.hp -= b * 2
            print(f'{defence.name}의 체력이 {defence.hp}가 되었다(hp{-b *2})')
        else:
            defence.hp -= b
            print(f'{defence.name}의 체력이 {defence.hp}가 되었다(hp{-b})')
        if defence.hp <= 0 :
            self.exp += 15
            print(f'{defence.name} 으앙주금')
            print(f'{self.name}이(가) + 15 경험치(현재 경험치 : {self.exp})')
        if self.exp >= 100:
            self.exp -= 100
            self.level += 1
```

```python
turn = 0
p1_name = input('플레이어 1의 이름을 정해주세요: ')
p2_name = input('플레이어 2의 이름을 정해주세요: ')
p1 = Pikachu(p1_name)
p2 = Pikachu(p2_name)
print('===========게임이 시작되었습니다================')
while True:
    print(f'==========={turn+1}턴================')
    a = input('qwer? : ')
    if a == 'q' or not a:
        if turn % 2:
            p2.body_attack(p1)
        else:
            p1.body_attack(p2)
    if a == 'r':
        if turn % 2:
            p2.thousond_volt(p1)
        else:
            p1.thousond_volt(p2)
    if p1.hp <= 0 or p2.hp <= 0:
        break
    turn += 1
```
